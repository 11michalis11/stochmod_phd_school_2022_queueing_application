```python
import csv

import matplotlib.pyplot as plt
import numpy as np
```

```python
value_losses = []
with open("data.txt", "r") as f:
    reader = csv.reader(f)
    data = list(reader)
```

```python
data[6:13]
```

```python
def get_value_of_measure(data, measure):
    raw_data = []
    for row in data:
        if measure in row[0]:
            raw_data.append(row[0])
    new_data = [float(i.replace(" ", "").split("|")[2]) for i in raw_data]
    return new_data
```

```python
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2, 3, figsize=(20, 10))

ax1.plot(get_value_of_measure(data, "entropy_loss"))
ax1.set_ylabel("entropy_loss")

ax2.plot(get_value_of_measure(data, "explained_variance"))
ax2.set_ylabel("explained_variance")

ax3.plot(get_value_of_measure(data, "learning_rate"))
ax3.set_ylabel("learning_rate")

ax4.plot(get_value_of_measure(data, "n_updates"))
ax4.set_ylabel("n_updates")

ax5.plot(get_value_of_measure(data, "policy_loss"))
ax5.set_ylabel("policy_loss")

ax6.plot(get_value_of_measure(data, "value_loss"))
ax6.set_ylabel("value_loss")
```

```python
plt.figure(figsize=(20, 10))
step = 150
demo = get_value_of_measure(data, "value_loss")[step:]
plt.plot(list(range(step, len(demo) + step)), demo)
plt.ylabel("value_loss")
```
